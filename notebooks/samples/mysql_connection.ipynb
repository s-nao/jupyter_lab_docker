```python
import pandas
import pymysql
```

```python
con = pymysql.connect(
        host='mysql',
        port=3306,
        db='world',
        user='username',
        password='password',
        charset='utf8'
    )
```

```python
with con.cursor() as cursor:
    sql = "select * from city;"
    cursor.execute(sql)

    # Select結果を取り出す
    for row in cursor.fetchall():
        print(row)
```
